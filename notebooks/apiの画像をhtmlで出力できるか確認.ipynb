```python
URL_DOMAIN="https://img.opf-dev.jp"
img_types=["osm",
                    "landsat8",
                    "landsat8-201808",
                    "true",
                    "natural",
                    "av2ori/band1",
                    "av2ori/band2",
                    "av2ori/band3",
                    "av2ori/band4"]

urls=["{}/{}/{}/{}/{}.png".format(URL_DOMAIN, img_type, z, x, y) for img_type in img_types]

imgs = np.array([io.imread(url) for url in urls])
```

```python
print(imgs.shape)
io.imshow(imgs[1])
```

```python
io.imshow_collection(imgs)
```

```python
io.imshow(imgs.reshape([-1, 256, 3]))
```

```python
#Grayscale
# color.rgb2gray
# 
# HSV : color.rgb2hsv
# YCbCr color.rgb2ycbcr
# YIQ color.rgb2yiq

img_gray_01 = color.rgb2gray(imgs[1])
# (0-1 -> 0-255)
img_gray = img_as_ubyte(img_gray_01)
io.imshow(img_gray)
img_gray
```

```python
# RGBYIQGrayScale
# YIQ 1 
img_yiq = color.rgb2yiq(imgs[1])
img_conb = np.concatenate((img_yiq[:, :, 1], img_yiq[:, :, 2]),axis=1)
io.imshow(img_conb)
img_conb
```

```python
img_conb2 = np.concatenate((img_yiq[:, :, 0], img_gray_01), axis=1) #
io.imshow(img_conb2)
```

```python
# RGBHSV
img_hsv = color.rgb2hsv(imgs[1])
io.imshow(img_hsv)
```

```python
im_conb_hsv = np.concatenate(( img_hsv[:, :, 0],  img_hsv[:, :, 1], img_hsv[:,:,2]), axis=1)
io.imshow(im_conb_hsv)
```
